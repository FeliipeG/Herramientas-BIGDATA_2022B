# Visualización de Datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Diagramas de Dispersión (scatterplot)

x = np.arange(2,20) # Asignamos valores iniciales
y = x**2
y2 = x**3
plt.scatter(x,y, color='black', marker='*', label='x**2') # Graficamos en color negro, que el marcador sea una estrella y el marcador se llame "x**2"
plt.scatter(x,y2, color='green', marker='+', label='x**3')
plt.plot(x,y2, color='pink', label='line') #.plot por defecto te grafica una linea. Agregar una gráfica encima de otra.

plt.title('Funciones Polinomicas', fontsize=20, color='red') # Graficamos el titulo en fuente 20 y color rojo.
plt.xlabel('x')  # Para mencionar el eje Y
plt.ylabel('y')  # Para mencionar el eje Y
plt.grid() # Para dejar las mallas
plt.legend() # Para mostras las leyendas de las variables "label"

plt.text(5, 4000, 'texto', color='blue')
```

```python
# Histogramas
mean = 5
std = 3
n = 10000
x = np.random.normal(mean, std, n) # Forma de generar 100 valores aleatorios que cumplan con el valor de la media, desviación estandar definida.
x

plt.hist(x, bins=20); # función bins para tomar los valores que crucen en 20

plt.title('Histograma Variable aleatoria', fontsize=20, color='red') # Graficamos el titulo en fuente 20 y color rojo.
plt.xlabel('x')  # Para mencionar el eje Y
plt.ylabel('Frecuencia Absoluta')  # Para mencionar el eje Y
plt.grid() # Para dejar las mallas
plt.legend() # Para mostras las leyendas de las variables "label"

plt.text(6, 1400, f'Mean: {mean} +/- {std}', color='blue')
```

```python
plt.boxplot(x, vert=False);  #Boxplot. vert= vertical = Es para colocarlo horizontal.
```

```python
#Subplots - Varias graficas en un lienzo
#Definir figura y ejes
fig, ax = plt.subplots(ncols=1, nrows=2, figsize=(5, 10), sharex=True)

ax[0].hist(x);
ax[1].boxplot(x, vert=False);
```

```python
# Realizaremos un diagrama de barras
#Barplot
# 1. definimos variables.
classes = ['gatos', 'perros', 'ratones']
totales = [20, 35, 12]

plt.bar(classes, totales)  #Si queremos la gráfica horizontal utilizamos .barh
```

```python
#Ahora hacer un grafico circular

plt.pie([1, 2, 3], totales)
#Pero mathplotlib no es eficiente para graficar pie
```

```python
# Procedimiento para realizar anotaciones sobre las barras, con Seaborn.
import seaborn as sns
# Para ello debemos crear un DataFrame
data = pd.DataFrame({
    'Animales': classes,
    'Totales' : totales
})

#data.head()

plots = sns.barplot(x='Animales', y='Totales', data=data)

for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.1f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=15, xytext=(0, 8),
                   textcoords='offset points')

```
