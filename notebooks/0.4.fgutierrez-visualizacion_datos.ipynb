# Visualización de Datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Diagramas de Dispersión (scatterplot)

x = np.arange(2,20) # Asignamos valores iniciales
y = x**2
y2 = x**3
plt.scatter(x,y, color='black', marker='*', label='x**2') # Graficamos en color negro, que el marcador sea una estrella y el marcador se llame "x**2"
plt.scatter(x,y2, color='green', marker='+', label='x**3')
plt.plot(x,y2, color='pink', label='line') #.plot por defecto te grafica una linea.

plt.title('Funciones Polinomicas', fontsize=20, color='red') # Graficamos el titulo en fuente 20 y color rojo.
plt.xlabel('x')  # Para mencionar el eje Y
plt.ylabel('y')  # Para mencionar el eje Y
plt.grid() # Para dejar las mallas
plt.legend() # Para mostras las leyendas de las variables "label"

plt.text(5, 4000, 'texto', color='blue')
```
